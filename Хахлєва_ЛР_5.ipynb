```python
import plotly_express as px
import pandas as pd
```

# 1. Для датасету px.data.iris() побудувати наступні графіки.  

Завантаження датасету Iris

```python
df = px.data.iris()
```

1. 3D-графік розкиду

```python
fig1 = px.scatter_3d(
    df,
    x='sepal_length',
    y='sepal_width',
    z='petal_length',
    color='species',
    title='3D Розподіл квіток Iris за довжиною чашолистка, шириною та довжиною пелюстки'
)

fig1.update_layout(title_x=0.5)
fig1.show()
```

2. Гістограма petal_width

```python
fig2 = px.histogram(
    df,
    x='petal_width',
    color='species',
    barmode='overlay',
    title='Розподіл petal_width для кожного виду Iris'
)

fig2.update_layout(title_x=0.5)
fig2.show()
```

3. Лінійний графік середнього sepal_length

```python
avg_df = df.groupby(['species', 'sepal_width']).mean(numeric_only=True).reset_index()

fig3 = px.line(
    avg_df,
    x='sepal_width',
    y='sepal_length',
    color='species',
    markers=True,
    title='Середня довжина чашолистка залежно від ширини для кожного виду Iris'
)

fig3.update_layout(title_x=0.5)
fig3.show()
```

# 2. Для датасету  px.data.gapminder() побудувати наступні

Завантаження датасету Gapminder

```python
gapminder = px.data.gapminder()
```

1. Гістограма ВВП на душу населення за 2007 рік (топ-10 країн)

```python
df = px.data.gapminder()
df_2007 = df[df['year'] == 2007]
top10_gdp = df_2007.nlargest(10, 'gdpPercap')

fig1 = px.bar(
    top10_gdp.sort_values('gdpPercap', ascending=False),
    x='country',
    y='gdpPercap',
    color='gdpPercap',
    color_continuous_scale='Viridis',
    text='gdpPercap',
    title='Топ-10 країн за ВВП на душу населення у 2007 році'
)

fig1.update_traces(texttemplate='%{text:.0f}', textposition='outside')
fig1.update_layout(
    title_x=0.5,
    xaxis_title='Країна',
    yaxis_title='ВВП на душу населення',
    uniformtext_minsize=8,
    uniformtext_mode='hide'
)
fig1.show()
```

2. Кругова діаграма ВВП країн Європи у 2007

```python
df_europe_2007 = df[(df['year'] == 2007) & (df['continent'] == 'Europe')]
top5_europe = df_europe_2007.nlargest(5, 'gdpPercap')

fig2 = px.pie(
    top5_europe,
    names='country',
    values='gdpPercap',
    title='Розподіл ВВП на душу населення (Європа, 2007)',
    hole=0.3
)

fig2.update_layout(title_x=0.5)
fig2.show()
```

3. Динаміка населення Канади і країн Океанії

```python
df_oceania_canada = df[(df['country'].isin(['Canada']) | (df['continent'] == 'Oceania'))]

fig3 = px.bar(
    df_oceania_canada,
    x='year',
    y='pop',
    color='country',
    barmode='group',
    title='Зміна населення Канади та країн Океанії (1952–2007)'
)

fig3.update_layout(
    xaxis_title='Рік',
    yaxis_title='Населення',
    title_x=0.5
)
fig3.show()
```
