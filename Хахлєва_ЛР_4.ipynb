# 2.Візуалізація даних

# Лінійний графік (відвідуваність Avila Adobe з часом)

```python
plt.figure(figsize=(16, 9))
plt.plot(df['month'], df['avila_adobe'], marker='o', linestyle='-', color='crimson')
plt.title('Відвідуваність Avila Adobe з часом')
plt.xlabel('Дата')
plt.ylabel('Кількість відвідувачів')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Boxplot (розподіл відвідувачів Chinese American Museum по роках)

```python
df['year'] = df['month'].dt.year
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(
    x='year',
    y='chinese_american_museum',
    data=df,
    palette='Set3',
    linewidth=2,
    fliersize=5
)

plt.title('Розподіл відвідувачів Chinese American Museum по роках', fontsize=16, weight='bold')
plt.xlabel('Рік', fontsize=12)
plt.ylabel('Кількість відвідувачів', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Heatmap (кореляція між музеями)

```python
museum_data = df.select_dtypes(include='number')

plt.figure(figsize=(12, 8))
sns.heatmap(
    museum_data.corr(),
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    linewidths=0.7,
    linecolor='white',
    square=True,
    cbar_kws={"shrink": 0.8}
)

plt.title('Кореляція між музеями', fontsize=16, weight='bold')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

# Histogram (розподіл кількості відвідувачів Gateway to Nature Center)

```python
plt.figure(figsize=(10, 6))
plt.hist(
    df['gateway_to_nature_center'].dropna(),
    bins=12,
    color='mediumseagreen',
    edgecolor='black',
    alpha=0.85
)

plt.title('Розподіл кількості відвідувачів Gateway to Nature Center', fontsize=14, weight='bold')
plt.xlabel('Кількість відвідувачів', fontsize=12)
plt.ylabel('Кількість місяців', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Barplot (середня кількість відвідувачів по музеях за всі роки)

```python
museum_means = df.select_dtypes(include='number').drop(columns='year').mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=museum_means.values, y=museum_means.index, palette='mako')
plt.title('Середня кількість відвідувачів по музеях (усі роки)', fontsize=14, weight='bold')
plt.xlabel('Середня кількість відвідувачів', fontsize=12)
plt.ylabel('Музей', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

У цій лабораторній роботі ми проаналізували дані про музеї за 2014–2021 роки: виконали очистку даних, перейменування колонок та виявили пропуски.
